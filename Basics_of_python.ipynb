## Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not.

```python
def is_prime(num):
    """Check if a number is prime."""
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Test examples
print(is_prime(2))   # True
print(is_prime(15))  # False
print(is_prime(17))  # True
```

## Exercise 2: Product of Random Numbers
Program that generates two random numbers and asks the user to enter the product.

```python
import random

def product_quiz():
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    print(f"What is the product of {a} and {b}?")
    user_answer = int(input("Enter your answer: "))
    if user_answer == a * b:
        print("Correct!")
    else:
        print(f"Incorrect. The correct answer is {a*b}.")

product_quiz()
```

## Exercise 3: Squares of Even Numbers (100–200)


```python
print("Squares of even numbers from 100 to 200:")
for i in range(100, 201):
    if i % 2 == 0:
        print(f"{i}^2 = {i**2}")
```

## Exercise 4: Word Counter
Count the number of words in a given text.

```python
from collections import Counter

def word_counter(text):
    words = text.split()
    return Counter(words)

# Example
input_text = "Python programming exercises. Python is easy programming language."
word_counts = word_counter(input_text)
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

## Exercise 5: Palindrome Checker
Function that checks if a string is a palindrome (ignoring spaces, punctuation, and capitalization).

```python
import string

def is_palindrome(s):
    # Normalize string: lowercase, remove spaces and punctuation
    translator = str.maketrans('', '', string.punctuation)
    cleaned = s.lower().translate(translator).replace(" ", "")
    return cleaned == cleaned[::-1]

# Test examples
print(is_palindrome("racecar"))       # True
print(is_palindrome("RaceCar"))       # True
print(is_palindrome("hello"))         # False
print(is_palindrome("A man a plan a canal Panama"))  # True
```
